```python
import pandas as pd
from sklearn import svm
from sklearn.metrics import roc_curve,auc,classification_report,confusion_matrix
from sklearn.model_selection import KFold,train_test_split
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
df=pd.read_csv("25%data.csv")
y=df["status_time"]
newdf = pd.DataFrame(scale(df), index=df.index, columns=df.columns)
newdf=newdf.round(3)
df=newdf
```

```python
df.drop("id",axis=1,inplace=True)

x=newdf[["LTV_dif","LTV_inc","LTV_time","FICO_orig_time","hpi_inc"]]
```

```python
kf=KFold(n_splits=5,random_state=2,shuffle=True)
```

```python
train_x, test_x, train_y, test_y=train_test_split(x, y, train_size=0.3)
```

```python
model = svm.SVC(kernel='poly',gamma=5,C=2) 
model.fit(train_x, train_y)
```

```python
model.score(test_x, test_y)
```

```python
pred_y = model.predict(test_x)
print(confusion_matrix(test_y, pred_y))
```

```python
print(classification_report(test_y, pred_y))
```
